```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt 
```

```python
 #importing the images
img1_path = '/content/gdrive/MyDrive/tenet/train/samantha/sam_2.jpeg'
img2_path = '/content/gdrive/MyDrive/tenet/train/samantha/sam_3.jpeg'
img3_path = '/content/gdrive/MyDrive/tenet/train/alia/ali_4.jpg'
#confirming the path of images
img1 = cv2.imread(img1_path)
img2 = cv2.imread(img2_path)
img3 = cv2.imread(img3_path)
```

```python
plt.imshow(img1[:, :, ::-1 ]) #setting value as -1 to maintain saturation
plt.show()
plt.imshow(img2[:, :, ::-1 ]) 
plt.show() 
plt.imshow(img3[:, :, ::-1 ]) 
plt.show()
```

```python
 #calling VGGFace
model_name = "VGG-Face"
model = DeepFace.build_model(model_name) 
```

```python
result = DeepFace.verify(img1_path,img2_path)#validate our images
DeepFace.verify("/content/gdrive/MyDrive/tenet/train/samantha/sam_2.jpeg","/content/gdrive/MyDrive/tenet/train/samantha/sam_3.jpeg")#generating result of comparison 
```

```python
#creating an object to analyze facial features
obj = DeepFace.analyze(img_path = "/content/gdrive/MyDrive/tenet/train/alia/ali_4.jpg", actions = ['age', 'gender', 'race', 'emotion'])
print(obj["age"]," years old ",obj["dominant_race"]," ",obj["dominant_emotion"]," ", obj["gender"])
```
